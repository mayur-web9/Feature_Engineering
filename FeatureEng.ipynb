5. What are continuous and categorical variables?

   => Continuous variables are numerical values that can take any value (e.g. height, weight).
   Categorical variables represent categories or groups     (e.g. gender, color).


6. How do we handle categorical variables in Machine Learning? What are the common techniques?

   => Categorical variables cannot be used directly in most Machine Learning models, so we convert them into numerical form.
   Common techniques are:-
   >Label Encoding – Assigns a number to each category       (e.g., Male = 0, Female = 1).

   >One-Hot Encoding – Creates separate columns for each category with 0 or 1 values.

   >Ordinal Encoding – Used when categories have an order   (e.g., Low, Medium, High).

16. What is an Optimizer? What are different types of optimizers? Explain each with an example.

    => An optimizer is a method used to reduce the loss (error) of a model by adjusting its parameters. Its main goal is to find the best values for weights so that the model performs better.

   > Common types of optimizers:
   Gradient Descent – Updates parameters step by step in the direction that reduces loss.
   Example: Used in linear regression to minimize error.

   > Stochastic Gradient Descent (SGD) – Updates parameters using one data point at a time. It is faster for large datasets.
   Example: Used in training neural networks.

   > Mini-batch Gradient Descent – Updates parameters using small batches of data instead of the full dataset.
   Example: Commonly used in deep learning models.

   > Adam (Adaptive Moment Estimation) – An advanced optimizer that adjusts the learning rate automatically.
   Example: Widely used in deep learning because it converges faster.

21. What is feature scaling? How does it help in Machine Learning?

    => Feature scaling is the process of bringing all numerical features to a similar range.

    > It helps because some algorithms (like KNN, SVM, and Gradient Descent) are sensitive to large values. If one feature has bigger numbers than others, it may dominate the model. Scaling improves accuracy and speeds up training.
